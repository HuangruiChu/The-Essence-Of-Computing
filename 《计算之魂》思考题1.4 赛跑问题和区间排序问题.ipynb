假定有25名短跑选手比赛争夺前三名，赛场上有五条赛道，一次可以有五名选手同时比赛。比赛并不计时，只看相应的名次。假如选手的发挥是稳定的，也就是说如果约翰比张三跑得快，张三比凯利跑得快，约翰一定比凯利跑得快。最少需要几组比赛才能决出前三名？

Q2. 区间排序。

如果有N个区间[l1, r1], [l2, r2], ... ,[lN, rN]，只要满足下面的条件我们就说这些区间是有序的：存在 [公式] ，其中i = 1, 2, ... N。

比如，[1, 4]、[2, 3]和[1.5, 2.5]是有序的，因为我们可以从这三个区间中选择1.1、2.1和2.2三个数。同时[2, 3]、[1, 4]和[1.5, 2.5]也是有序的，因为我们可以选择2.1、2.2和2.4。但是，[1, 2]、[2.7, 3.5]和[1.5, 2.5]不是有序的。

对于任意一组区间，如何将它们进行排序？

思路：归并排序的合并步骤

第一步，将25名跑步者分成5组（A组、B组、C组、D组和E组），每组5个选手。

A组的5名选手速度从快到慢排序依次为A1、A2、A3、A4、A5；

B组的5名选手速度从快到慢排序依次为B1、B2、B3、B4、B5；

C组的5名选手速度从快到慢排序依次为C1、C2、C3、C4、C5；

D组的5名选手速度从快到慢排序依次为D1、D2、D3、D4、D5；

E组的5名选手速度从快到慢排序依次为E1、E2、E3、E4、E5；

第二步，将A1,B1,C1,D1,E1 比赛，得到整体第一名。不失性地认为A1是整体第一名。

第三步，A2,A3,B1,B2,C1 比赛，可以得出整体 第二 第三名

“

根据香农对信息的描述——信息是可以消除不确定性的东西。为了满足题目“最少”的要求，我们需要使用尽可能少得信息来消除不确定性。看一下5组中剩下的15名选手，哪些选手的信息载荷更大呢？对于选择跑得更快这一目的，每组的第一名，其信息载荷更大。因为当一个选手X的速度比另一组中速度最快的选手还要快时，X一定比那一组中的第二名和第三名的速度要快。

所以，我们用有限的资源——5条跑道，去测试5组中速度最快的5个人，也就是A1、B1、C1、D1和E1。

不失一般性，我们可以假设最终5人的成绩为A1>B1>C1>D1>E1。

通过上面的结果，我们可以将一些选手排除在最终前三名的范围之外。

首先，D组和E组剩余的所有选手都不会进入前三名，因为排在D1前面的选手至少有3名；

其次，C2和C3也不会进入前三名，因为A1>B1>C1>C2>C3；同理，B2也不在前三名中；

这是我们通过测试5组中最快选手能够得到的明确的排除信息。同时，我们还能够知道，A1是所有选手中跑得最快的选手，自然进入前三名。

此时，前三名还有两个名额将在A2、A3、B1、B2和C3这5名选手中产生，让这5人再跑一次，便能得到最终的结论。

综上，最少需要7组比赛才能决出前三名。
